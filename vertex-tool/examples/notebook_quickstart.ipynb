# Claude on Vertex AI - Workbench quickstart

Run this notebook on a Vertex AI Workbench instance after installing the offline bundle (`scripts/install.sh`).

Authentication uses Application Default Credentials supplied by the Workbench service account, so no API key is needed.

```python
import os
os.environ.setdefault('ANTHROPIC_VERTEX_PROJECT_ID', 'your-gcp-project')
os.environ.setdefault('CLOUD_ML_REGION', 'us-east5')
```

```python
from claude_vertex_tool import ClaudeVertexClient

client = ClaudeVertexClient()
print(client.send('Vertex AI 上の Claude を1文で説明して。'))
```

```python
for chunk in client.stream('箇条書きで3点、Workbench から Claude を呼ぶ利点を。'):
    print(chunk, end='', flush=True)
```

```python
messages = [
    {'role': 'user', 'content': '次の Python コードを最適化してください: sum([i*i for i in range(10000)])'},
]
print(client.send(messages, system='You are a senior Python engineer. Reply in Japanese.'))
```
